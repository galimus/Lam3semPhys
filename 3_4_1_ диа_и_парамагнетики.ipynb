# **1)Теория**

В данной работе используется метод Гюи . 
Сила действующая на образец со стороны магнитного поля :
$$
F_M=(\frac{\partial W_M}{\partial x})_I ;
\text где W_M(x) -магнитная\,энергия\,системы \, при\, I=const\,(то\,есть\,при\,B_o=const)\,в\,зависимости\,от\,x
$$


Магнитная энергия ,для объемной плотности может быть рассчитана по формуле:
$$ 
W_M=\frac{1}{2\mu_0}\int\frac{B^2}{\mu}dV
$$
где интеграл распространен на все пространство

изменение магнитной энергии при смещение dx:
$$
dW_m(\Delta x)\approx \frac{B^2_2}{2\mu\mu_0}Sdx-\frac{B^2_3}{2\mu_0}Sdx=(\mu-1)\frac{B^2_0}{2\mu_0}Sdx
$$
Следовательно ,искомая сила равна 
$$
F_M=(\frac{\partial W_M}{\partial x})_{B_0}\approx \chi\frac{B^2_0}{2\mu_0}S
$$
знак силы зависит от знака восприимчивости $$\chi =\mu -1 ;$$ парамагнетики  $$\chi >0 $$ втягиваются в зазор электромагнита ,а диамагнетики $$\chi<0$$ выталкиваются из него .Таким образом ,измерив силу,действующую на образец в магнитном поле,можно рассчитать магнитную восприимчивость образца.
